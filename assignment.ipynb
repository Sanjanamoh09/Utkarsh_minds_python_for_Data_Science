```python
# Wilcoxon Test:

# 1. Perform a Wilcoxon rank-sum test to compare the median of the first dataset is different from hypothetical median.
# 2. Print the test statistic and p-value, and interpret the result.

from scipy.stats import wilcoxon

'''
H0: The median of the dataset is not significantly different from the hypothetical median.
Ha: The median of the dataset is significantly different from the hypothetical median.
'''

#x = hypothetical median
x = 45
t_stat, pvalue = wilcoxon(df['Random numbers - I'] - x, zero_method = 'wilcox')

print("t_statistic: ",t_stat,"\nP-value: ",pvalue)

alpha = 0.05  # significance level
if pvalue < alpha:
    print("Reject the null hypothesis: The mean of the dataset is significantly different from the hypothetical median.")
else:
    print("Accept null hypothesis: There is no significant difference between the medain of the dataset and the hypothetical median for Dataset 1")
```

```python
# Pearson Goodness of Fit Test:

# 1. Perform a Pearson goodness of fit test to check if the distribution of numbers in the first dataset fits a uniform distribution.
# 2. Print the test statistic and p-value, and interpret the result.

from scipy.stats import chisquare

'''
H0: The data follows a uniform distribution.
Ha: The data doesn't follow a uniform distribution.
'''

chi_stat, pvalue = chisquare(df['Random numbers - I'].value_counts())
print("Test statistic: ",chi_stat,"\nP-value: ",pvalue)

alpha = 0.05  # significance level
if pvalue < alpha:
    print("Reject the null hypothesis: The data doesn't follow a uniform distribution.")
else:
    print("Accept null hypothesis: The data follows a uniform distribution.")
```
