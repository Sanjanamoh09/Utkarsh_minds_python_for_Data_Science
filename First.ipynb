```python
#HYPOTHESIS TEST - BINOMIAL TEST
'''
sample = 46
male = 34
female = 12

H0 = null hypothesis - there is no difference in number of male and female in the population.
Ha = alternative hypothesis - there is a difference in number of male and female in the population.

Two-sided/tailed test: non-directional (equal or not equal)
One-sided/tailed test: directional (less or greater)

Goal: reject the null hypothesis or accept the alternative hypothesis (no evidence to reject H0).
'''
```

```python
from scipy.stats import binom_test

pvalue=binom_test(x=34, n=46, p=0.5)
if pvalue<=0.05:
  print("Pvalue = ", pvalue)
  print("As pvalue <= 0.05, reject H0. Thus, there is a difference in number of male and femail in the population.")
  print("Our result is significant.\n")
else:
  print("Accept H0")
  print("Our result is not significant.")

```

```python
#T-TEST
data=pd.read_csv("GSS2012a.csv")
scalevar=data['age'].dropna()
scalevar=scalevar.replace({'89 OR OLDER': 90})
scalevar=pd.to_numeric(scalevar)
```

```python
from scipy.stats import ttest_1samp
'''
RQ - is the average age of the population is 45?

H0 = the average age of the population is 45.
Ha = the average age of the population is different.

p<0.05, H0 is rejected and average age is not 45.

effect size = cohen 'd


'''

ttest_1samp(scalevar, popmean = 45)
```

```python
statistic=8.022891023250635
pvalue=1.7566438579204372e-15
```

```python
#false_positive_percent = less than 1%


#As p<0.05, H0 is rejected and average age is not 45.
```

```python
m=scalevar.mean()
sd=scalevar.std()

#Effect size - CALCULATED ONLY WHEN THE PVALUE IS LESS THAN 0.05

#Ha -  the mean of the population is different from the hypothetical mean

meanpop = 80
temp = (m - meanpop)/sd
cohen_d = abs(temp * (2**0.5))

if cohen_d<0.2:
  print("Negligible")
elif cohen_d<0.5:
  print("Small")
elif cohen_d<0.8:
  print("Medium")
else:
  print("Large")
```

```python
#Population difference is large from the hypothetical population as the cohen_d interpretation comes out to be "LARGE"
```

```python
from scipy.stats import wilcoxon
x = 2.5
wilcoxon(ord - x, zero_method = 'wilcox')
```

```python
pvalue = 8.376600483105338e-142

#Ha accepted.

#Median of the population don't find accounting as scientific.

```

```python
from scipy.stats import norm

#pvalue < 0.05
# Ha = the median value of the population is different from x (hypothetical median).

# if pvalue < 0.05, effect size - negligible, small, medium, large.

# ROSENTHAL CORRELATION COEFFICIENT.

z=norm.ppf(pvalue/2)
n=len(ord)
r_ros = abs(z/n **0.5)

if r_ros<0.2:
  print("very small")
elif r_ros<0.4:
  print("Small")
elif r_ros<0.6:
  print("Medium")
elif r_ros<0.8:
  print("large")
else:
  print("very large")
  # A one-sample wilcoxon signed-rank test indicates that median of the population is different from x = 2.5 (p<0.05) and the rosenthal correlation coefficient indicates the difference is very large.
```

```python
# Hypothesis Testing:

# 1. Formulate null and alternative hypotheses to test if the mean of the dataset is significantly different from a hypothetical mean.

from scipy import stats

'''
H0: The mean of the dataset is not significantly different from the hypothetical mean.
Ha: The mean of the dataset is significantly different from the hypothetical mean.
'''

#x = hypothetical mean
x=50
sample_mean = mean2
t_stat, pvalue=stats.ttest_1samp(df1['Random Integers - II'], x)


alpha = 0.05  # significance level
if pvalue < alpha:
    print("Reject the null hypothesis: The mean of the dataset is significantly different from the hypothetical mean.")
else:
    print("Accept null hypothesis: There is no significant difference between the mean of the dataset and the hypothetical mean for Dataset 2")
```

```python
# Pearson Goodness of Fit Test:

# 1. Perform a Pearson goodness of fit test to check if the distribution of numbers in the first dataset fits a uniform distribution.
# 2. Print the test statistic and p-value, and interpret the result.

from scipy.stats import chisquare

'''
H0: The data doesn't follow a uniform distribution.
Ha: The data follows a uniform distribution.
'''

chi_stat, pvalue = chisquare(df['Random numbers - I'].value_counts())
print("Test statistic: ",chi_stat,"\nP-value: ",pvalue)

alpha = 0.05  # significance level
if pvalue < alpha:
    print("Reject the null hypothesis: The data doesn't follow a uniform distribution.")
else:
    print("Accept null hypothesis: The data follows a uniform distribution.")
```
